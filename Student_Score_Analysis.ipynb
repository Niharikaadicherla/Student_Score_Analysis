```python
#  want to check gender distribtion 

plt.figure(figsize = (4,4))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore" :"mean", "ReadingScore": "mean" , "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize = (4,6))
sns.heatmap(gb, annot = True)
plt.title("relationship between  parent's education and scores")
plt.show()
```

Based on the analysis of the chart, it is evident that parental education levels have a significant positive impact on their children's scores. This correlation suggests that higher levels of parental education are associated with better academic performance by their children.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore" :"mean", "ReadingScore": "mean" , "WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,6))
sns.barplot(gb1)
plt.title("relationship between parent's marital status and scores")
plt.show()
```

The chart shows that the marital status of students' parents doesn’t seem to affect their academic performance much. Whether students’ parents are married or single, or in any other situation, it doesn’t make a big difference in their scores. This means that students’ scores are similar regardless of their parents’ marital status.

```python
sns.violinplot(x='WklyStudyHours', y='TransportMeans', data=df, color = 'crimson')
plt.figure(figsize = (12,6))
plt.title("study hours based on transport")
```

The violin plot suggests that students using the school bus generally have higher weekly study hours compared to those using private transport. This is indicated by a wider distribution and higher density of study hours for school bus riders. The differences in study hours could be influenced by factors such as extended commuting time, consistent routines, and access to study resources. Further investigation is needed to understand the reasons behind these differences and confirm the trends observed in the plot.

```python
# Assuming df is your DataFrame and contains the columns 'WritingScore', 'ReadingScore', and 'MathScore'

# Create a figure and axes
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Plot distribution of WritingScore
sns.histplot(df['WritingScore'], kde=True, ax=ax[0], color='blue')
ax[0].set_title('Distribution of WritingScore')

# Plot distribution of ReadingScore
sns.histplot(df['ReadingScore'], kde=True, ax=ax[1], color='green')
ax[1].set_title('Distribution of ReadingScore')

# Plot distribution of MathScore
sns.histplot(df['MathScore'], kde=True, ax=ax[2], color='red')
ax[2].set_title('Distribution of MathScore')

plt.tight_layout()
plt.show()
```

From Histogram Provides insight into the frequency distribution of each score, allowing for visual assessment of patterns, central tendencies, and spread of the data.
KDE Curve helps us to understand the underlying distribution of the data beyond the discrete bins of the histogram, showing where data points are more densely concentrated.
